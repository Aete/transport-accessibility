```python
import geopandas as gpd
import pandas as pd
import numpy as np
```

```python
gdf_building = gpd.read_file('../gangnamGu-building.geojson', encoding='utf8', dtype={'USABILITY':str})
```

```python
gdf_building.head()
```

```python
gdf_building.columns
```

```python
gdf_building['USABILITY'].unique()
```

```python
# land use code
# 01000 단독주택
# 01001 단독주택
# 01002 다중주택
# 01003 다가구주택
# 01004 공관
# 02000 공동주택
# 02001 아파트
# 02002 연립주택
# 02003 다세대주택
# 02007 기숙사
# 10202 오피스텔

residential_code = ['01000', 
                   '01001',
                   '01002',
                   '01003',
                   '01004',
                   '02000',
                   '02001',
                   '02002',
                   '02003',
                   '02007',
                   '10202']
```

```python
# filtering the dataset by the land use code
gdf_res_building = gdf_building.loc[gdf_building['USABILITY'].isin(residential_code)]
```

```python
gdf_res_building.columns
```

```python
gdf_res_building = gdf_res_building.loc[:,['USABILITY',
                                           'HEIGHT',
                                           'PNU',
                                           'GRND_FLR',
                                           'UGRND_FLR',
                                           'DONG_NM',
                                           'COL_ADM_SE',
                                           'UFID',
                                           'BLDG_PNU',
                                           'BD_MGT_SN',
                                           'ARCHAREA',
                                           'TOTALAREA',
                                           'BC_RAT',
                                           'VL_RAT',
                                           'geometry']]
```

```python
gdf_res_building.head()
```

```python
gdf_res_building.to_file('../gangnamGu-building-residential.geojson', driver='GeoJSON', encoding='utf8')
```
