```python
import numpy as np
import pandas as pd
import geopandas as gpd
```

```python
gdf_buildings = gpd.read_file('../gangnamGu-building-residential.geojson')
```

```python
gdf_buildings.head()
```

```python
gdf_hexbin = gpd.read_file('../hexagon_res_9_400m_count/hexagon_with_data.geojson')
```

```python
gdf_hexbin.head()
```

```python
gdf_buildings = gdf_buildings.loc[:,['UFID','geometry']]
gdf_hexbin = gdf_hexbin.loc[:,['index','geometry']]
```

Here, I splitted the residential buildings into segments by the hexagons

```python
intersections = gpd.overlay(gdf_buildings, gdf_hexbin, how='intersection')
```
